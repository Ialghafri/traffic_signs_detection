```python
import torchvision.models as models
from PIL import Image
import time
import cv2
import random
import matplotlib.pyplot as plt
import os
from ultralytics import YOLO
```

```python
images = 'traffic_sign_inferencing'
```

```python
num_samples = 9
image_files = os.listdir(images)

# Randomly select num_samples images
rand_images = random.sample(image_files, num_samples)

fig, axes = plt.subplots(3, 3, figsize=(11, 11))

for i in range(num_samples):
    image = rand_images[i]
    ax = axes[i // 3, i % 3]
    ax.imshow(plt.imread(os.path.join(images, image)))
    ax.set_title(f'Image {i+1}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

YOLOmodel = YOLO('server/best.pt')

start_time = time.time()

image2 = 'inferences/images-2.jpeg'
img2 = mpimg.imread(image2)
prediction2 = YOLOmodel.predict(source= 'traffic_sign_inferencing', save=True)

end_time = time.time()

pt_total_time = end_time - start_time
pt_average_time = pt_total_time / len(images)

print(f"Total inference time: {pt_total_time:.4f} seconds")
print(f"Average inference time per image: {pt_average_time:.4f} second")
```
